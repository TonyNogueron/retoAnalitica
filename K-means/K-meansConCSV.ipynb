**Dataset**

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

tabla =  pd.read_csv("/content/drive/MyDrive/Universidad/Segundo Semestre/Semana Tec Ramona Analítica/Mall_Customers.csv", low_memory=False)
tabla = tabla[['Age','Spending Score (1-100)']]

age = tabla['Age']
score = tabla['Spending Score (1-100)']

# plot
plt.scatter(age, score)
plt.show()
x = tabla.iloc[:,:]
```

**K-means**

```python
from sklearn.cluster import KMeans

kmeans = KMeans(6)
kmeans.fit(x)
identified_clusters = kmeans.fit_predict(x)
```

**Visualization**

```python
data_with_clusters = tabla.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Age'],data_with_clusters['Spending Score (1-100)'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

**Elbow method**

```python
# calculate distortion for a range of number of cluster
wcss=[]
for i in range(1,7):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
